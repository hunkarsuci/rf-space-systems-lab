````python
# Floating-point numbers 

# A floating-point number contains a decimal point 
# or is written using scientific notation 

frequency_hz = 1000.0
distance_m = 25.5
power_w = 10.0

print(frequency_hz)
print(distance_m)
print(power_w)

print(type(frequency_hz))
print(type(distance_m))
print(type(power_w))
````

## Engineering quantities and units 

A numerical value by itself is usually incomplete in engineering.

For example, the number:

```text
10

````python
# Store and print engineering quantities

frequency_hz = 2_400_000_000.0
distance_m = 1_000.0
power_w = 5.0
time_s = 0.25

print("Frequency:", frequency_hz, "Hz")
print("Distance:", distance_m, "m")
print("Power:", power_w, "W")
print("Time:", time_s, "s")
````
